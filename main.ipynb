## 2.1 动态提问加入记忆库

```python
import os
from dotenv import load_dotenv
from llm.zhipu import zhipuLLM
from embedding.zhipu import zhipuEmb
from memory.memory import JSONMemoryBank

# 1. 加载模型
load_dotenv()
api_key = os.getenv("ZHIPU_API_KEY")
llm = zhipuLLM(model_name="glm-4", api_key=api_key)
emb = zhipuEmb(model_name="embedding-2", api_key=api_key)

# 2. 初始化记忆库
memory_bank = JSONMemoryBank(embedding_model=emb, file_path="my_memory.json")
memory_bank.clear()  # 清空之前的记忆，方便测试

# 3. 测试存储功能 (原有静态记忆)
print("\n--- 开始测试记忆存储 ---")

memory_bank.add(
    text="我叫豆包，是一个AI实验助手，我正在帮用户开发Agent。",
    metadata={"type": "user_profile", "source": "user_input"}
)

memory_bank.add(
    text="2026年的今天是4月2日，我们正在做一个基于JSON的记忆库。",
    metadata={"type": "fact", "source": "experiment"}
)

# ====================== 新增部分开始 ======================
# 4. 动态提问LLM并存储
print("\n--- 动态提问LLM并存储 ---")

# 4.1 定义问题
query = "林黛玉和贾宝玉是什么关系？"
print(f"你问: {query}")

# 4.2 调用LLM获取回答
llm_response = llm.predict(query)
print(f"LLM答: {llm_response}")

# 4.3 将问答对存入记忆库
memory_bank.add(
    text=f"用户问: {query}\nAgent答: {llm_response}",
    metadata={"type": "dialogue", "source": "llm_chat"}
)
# ====================== 新增部分结束 ======================

# 5. 打印一下所有记忆看看
print("\n--- 当前记忆库内容 ---")
all_mem = memory_bank.get_all()
for mem in all_mem:
    print(f"[{mem['formatted_time']}] {mem['text'][:30]}...")
```
